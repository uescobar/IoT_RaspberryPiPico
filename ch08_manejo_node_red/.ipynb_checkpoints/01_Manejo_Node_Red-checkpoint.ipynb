Entre los dos componentes viaja información a través de su unión.

Lo que viaja es un mensaje, que se puede interpretar como un objeto que está viajando por el canal.

```
// Circula un mensaje por el flujo
// el formato del mensaje es un JSON

{parametro1:valor1, parametro2: valor2, parametro3:valor3}

// así pueden exitir más parámetros, y expresado en otra forma más corta

{p1:v1, p2:v2, p3:v3}

// otro ejemplo, los nombres de lo parametros siempre van a ser string
// el valor del parámetro depende de lo que deseemos comunicar, entero, flotante, lista, diccionario, JSON, etc.

{"Nombre":"Jorge", "Edad":39, "notas":{"Virtualizacion":10, "Programacion":8}}

// se puede trabajar con los parámetros que se crean necesarios, pero hay un parámetro particular, se llama payload,
// que contiene la información, los demás parámetros suelen ser indicativos, pero para objetos predefinidos,
// el parámetro de los datos normalmente se llama payload.

{"Nombre":"Jorge", "Edad":39, "Payload":"Data"}

// Por ejemplo, si uso mqtt para publicar tengo que publicar un JSON, y si me suscribo recibo un JSON,
// los elementos que ya estan predefinidos trabajan con el parámetro payload

// lo importante es entender el formato JSON, 

// entonces, loq ue viaja es un elemento en formato JSON

// mensaje msg ==> {p1:v1, p2:v2, p3:v3}

En las propiedades de los elementos de node-red se puede acceder a las propiedades de la siguiente manera

msg.p1 
msg.Edad
msg.payload

```

## MQTT suscriptor en node-red


Existen dos componentes de MQTT, 

* mqtt in, que se conecta a un broker y se subscribe a mensajes de topicos específicos
* mqtt out, que se conecta a un broker mqtt y publica mensajes.

NODERED NO IMPLEMENTA UN SERVIDOR BROKER MQTT, sino simplemente, nodered es un cliente de mqtt, por lo que hayq ue asegurarse de que el servidor broker de MQTT se encuentre funcionando.

### Ejecutar Select desde node a la Base de datos.

De acuerdo a la documentación de node-red, se especifica los siguiente:


*This node uses the query operation against the configured database. This does allow both INSERTS and DELETES. By its very nature it allows SQL injection... so be careful out there...*

**msg.topic** must hold the query for the database, and the result is returned in msg.payload.

**msg.payload** can contain an array of values to bind to the topic.

Así que la consulta debe estar situada en el atributo **topic** y los resultados de la consulta se regresan en **payload**.

Pero también en **payload** puede tener asociado un arreglo de valores que va a usar la consulta que está en **topic**.

El primer ejemplo a realizar será un **SELECT**, para lo cual se requiere el elemento **inject** que se había denominado previamente como **prueba:hola** y además del componente de **mysql** se requiere otro componente **debug** (que solo va a mostrar en la salida el **payload**) y se unen en ese orden.

![88.png](img/88.png)

La instrucción que va a realizar la base de datos debe ir en el elemento **topic** de la entrada y no tiene data en el payload, así que lo que tenga payload no interesa en este momento. Para ubicar los elementos en los campos corresponientes se debe dar doble click en el elemento **inject** **prueba:hola**

***
```SQL
SELECT nombre, edad FROM tabla1;
```
***

![89.png](img/89.png)

Para ver la salida del elemento, debemos habilitar el panel debug, y dams click en deploy

![90.png](img/90.png)

Para ejecutar la consulta debemos dar click en el recuadro a la izquierda del elemento **inject**

### Insertar datos desde node a la Base de datos.

Recordando brevemente, el elemento **prueba** es un elemento que esta suscrito a mqtt y por lo tanto desde otra consola se va a enviar la información que será agregada a la base de datos.

Para realizar lo siguiente se va a retulizar lo que se tiene en el elemento **prueba** y unirlo con el elemento de mysql. Para lo cual se va a eliminar el elemento **inject** en el que se tenia el query de select.

![93.png](img/93.png)

Sin embargo, lo que se recibe es un string que debe ser convertido a json, por lo que usaremos un elemento **json** simplemente arrastrando y soltando sobre la línea que une los elementos **prueba** y el elemento **mysql** (Servidor OCI) .

![94.png](img/94.png)

Para poner la salida del **json** y formar correctamente la sintaxis para comunicarse con la base de datos, se requiere modificar esa salida mediante un elemento **function** que a la vez va a usar código **javascript**, por lo que después del elemento json, colocamos un elemento **function**.

![95.png](img/95.png)

Sin embargo, para hacer pruesbas, se va a simplificar momentaneamente el flujo, tal como se ve a continuación, se eliminan elementos innecesarios, se saca momentáneamente el elemento de base de datos **mysql** (Servidor OCI) del flujo, y la salida de **function 1** se va a **debug 2**, para realizar pruebas y verificar que el formato de salida es el que buscamos.

![96.png](img/96.png)

Ahora que se sabe que funciona el código, es momento de realizar cambios para enviar la información de otra forma, ya que la forma en la que está programado solo esta enviando string, que implícitamente se convierten a byte arrays.


**08_ejemplo_01_mqtt_publicador_mensajes.py**


***
```Python

import network
import utime
import umqtt.simple as mqtt



### SECCION DE CONEXION CON wifi
wf = network.WLAN(network.STA_IF)
wf.active(True)

wf.connect('ISSD', 'password red')


# la conexión puede demorar unos segundos,
# Creamos un ciclo que termina cuando se ha realizado la conexión
while not wf.isconnected():
    print(".")
    utime.sleep(1) # cada segundo se imprime un punto

# al salir, imprimimos
print("Conectado al WiFi")
print(wf.ifconfig())

### SECCION DE CONEXION CON mqtt
#el id "pico1" debe ser único, si hay mas picos, estos ids deben ser diferentes
#la ip del broker 159.54.134.99
c=mqtt.MQTTClient("pico1", "159.54.134.99")

#conectamos
c.connect()
#ciclamos la recepcion de mensajes
print("Ejecutando el suscriptor")
continuar = True
while continuar:
    a = input('Escriba el mensaje: ')
    b = [7,16] ## cambio esta línea
    if a == 'z':
        continuar = False
    else:
        c.publish('prueba', bytearray(b))

c.disconnect()
print('Fin del programa')

```
***

Antes de ejecutar el programa, se debe cambiar la forma en la que la información del objeto mqtt será presentada a la salida, para lo cual hay que dar doble click sobre el nodo **mqtt prueba** y en **Outuput** seleccionar **auto-detect (parsed JSON object, string or buffer)**, dar click en **Done** y posteriormente en **Deploy**.

![104.png](img/104.png)

Sin embargo, se debe tomar en cuenta que se está interpretando su valor, ya que realmente están llegando a mqtt un objeto con valores binarios que es interpretado como código hexadecimal que se muestran cuando se despliega el contenido completo del objeto recibido.

![105.png](img/105.png)

Sin embargo, cuando los valores sobre pasan el valor 256, se envían datos incorrectos porque el valor excedería lo que se puede almacenar en un byte. En el ejemplo se envío un valor de 257 y lo representa como 1.

![106.png](img/106.png)

Para solucionar el problema se puede enviar en formato JSON. Por lo que el código sería de la siguiente manera.

**08_ejemplo_01_mqtt_publicador_mensajes.py**


***
```Python
import network
import utime
import umqtt.simple as mqtt

import json  ## cambio esta linea



### SECCION DE CONEXION CON wifi
wf = network.WLAN(network.STA_IF)
wf.active(True)

wf.connect('ISSD', 'password red')


# la conexión puede demorar unos segundos,
# Creamos un ciclo que termina cuando se ha realizado la conexión
while not wf.isconnected():
    print(".")
    utime.sleep(1) # cada segundo se imprime un punto

# al salir, imprimimos
print("Conectado al WiFi")
print(wf.ifconfig())

### SECCION DE CONEXION CON mqtt
#el id "pico1" debe ser único, si hay mas picos, estos ids deben ser diferentes
#la ip del broker 159.54.134.99
c=mqtt.MQTTClient("pico1", "159.54.134.99")

#conectamos
c.connect()
#ciclamos la recepcion de mensajes
print("Ejecutando el suscriptor")
b = {"v1":0, "v2":0, "v3":0} ## cambio esta línea, solamente para definir el objeto diccionario
continuar = True
while continuar:
    a = input('Escriba el mensaje: ')
    # actualizamos los valores de b
    b["v1"] = 7 ## cambio esta línea
    b["v2"] = 256 ## cambio esta línea
    b["v3"] = 512 ## cambio esta línea
    if a == 'z':
        continuar = False
    else:
        c.publish('prueba', json.dumps(b)) ## cambio esta linea

c.disconnect()
print('Fin del programa')

```
***

Y ahora se realia la prueba, y se puede observar que ya se recibe el formato JSON, con los valores adecuados sin modificar.

```PYTHON

import network
import utime
import umqtt.simple as mqtt

import json  ## cambio esta linea

import random ## AGREGO ESTA LINEA

### SECCION DE CONEXION CON wifi
wf = network.WLAN(network.STA_IF)
wf.active(True)

#wf.connect('ISSD', 'password red')
# wf.connect('IZZI-0761', 'C85261810761')
wf.connect('pilotAcademic', 'C85261810761')

# la conexión puede demorar unos segundos,
# Creamos un ciclo que termina cuando se ha realizado la conexión
while not wf.isconnected():
    print(".")
    utime.sleep(1) # cada segundo se imprime un punto

# al salir, imprimimos
print("Conectado al WiFi")
print(wf.ifconfig())

### SECCION DE CONEXION CON mqtt
#el id "pico1" debe ser único, si hay mas picos, estos ids deben ser diferentes
#la ip del broker 159.54.134.99
c=mqtt.MQTTClient("pico1", "159.54.134.99")

#conectamos
c.connect()
#ciclamos la recepcion de mensajes
print("Ejecutando el suscriptor")
b = {"v1":0, "v2":0, "v3":0} ## cambio esta línea, solamente para definir el objeto diccionario
continuar = True
while continuar:

    # actualizamos los valores de b
    b["v1"] = random.randint(0,50) ## CAMBIO ESTA LINEA
    b["v2"] = random.randint(0,200) ## CAMBIO ESTA LINEA
    b["v3"] = random.randint(0,500) ## CAMBIO ESTA LINEA
    c.publish('prueba', json.dumps(b)) ## cambio esta linea
    print("dato enviado") # CAMBIO ESTA LINEA
    utime.sleep(5) # CAMBIO ESTA LINEA, se ejecuta el ciclo cada 5 segundos
    
c.disconnect()
print('Fin del programa')


```
